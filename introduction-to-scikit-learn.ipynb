### 0. An end-to-end Scikit-Learn workflow

### 1. Get the data ready

```python
import pandas as pd
import numpy as np
heart_disease = pd.read_csv("../assigments_env/heart-disease.csv")
heart_disease
```

```python
# Create x (features matrix)
X = heart_disease.drop("target", axis=1)

# Create y (labels)
y = heart_disease["target"]
```

### 2. Choose the right model and hyperparameters

```python
from sklearn.ensemble import RandomForestClassifier  
clf = RandomForestClassifier(n_estimators=100)

# We'll keep the default hyperparameters
clf.get_params()
```

### 3. Fit the model to the data

```python
from sklearn.model_selection import train_test_split

# We are gonna use 80% of the data to train and the rest to test.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Asking to the classification model to find the patterns in the training data
clf.fit(X_train, y_train)
```

```python
X_train
```

```python
X_test
```

```python
# Make a prediction with test data
y_preds = clf.predict(X_test) 
y_preds
```

```python
y_test
```

### 4. Evaluate the model on the training data and test data

```python
# Returns the mean accuracy on the given test data and labels
clf.score(X_train, y_train)
```

```python
 clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Shows us some classification metrics that compare the test labels to the predicctions labels
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

 ### 5. Improve a model

```python
# Try different amount of n_stimators
np.random.seed(42)
for i in range(10, 100, 10):
        print(f"Trying model with {i} estimators")
        clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
        print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
        print("")
```

 ### 6. Save the model and load it

```python
import pickle

pickle.dump(clf, open("random_forst_model_1.pkl", "wb")) # wb = write binaries
```

```python
loaded_model = pickle.load(open("random_forst_model_1.pkl", "rb")) # rb = read binaries
loaded_model.score(X_test, y_test)
```
